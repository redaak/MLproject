```python
def g(x):
    return x**4-4*x**2+5
def dg(x):
    return 4*x**3-8*x
```

```python
x_2= np.linspace(-2,2,1000)
```

```python
# gradient decisent function
def gradient_decsent(derivative_func,star_x,step_multiplier=0.01,max_iter=500):
    previouse_x=0
    step_size=star_x
    xList=[star_x]
    xSlope=[derivative_func(star_x)]
    for n in range(max_iter) :
        previouse_x=star_x
        gradient = derivative_func(previouse_x)
        star_x= previouse_x - step_multiplier * gradient
        xList.append(star_x)
        xSlope.append(derivative_func(star_x))
    return star_x , xList , xSlope
    
```

```python
results = gradient_decsent(derivative_func=dg,star_x=1.9,step_multiplier=0.01,max_iter=500)
```

# Data viz with 3D charts
## $$ f(x,y)= \frac{1}{3^{-x^2-y^2}+1}$$
## $$ \frac{\partial f}{\partial x}= \frac{2x\ln(3)\cdot 3^{-x^2-y^2}}{\left(3^{-x^2-y^2}+1\right)^2}$$
## $$ \frac{\partial f}{\partial y}= \frac{2y\ln(3)\cdot 3^{-x^2-y^2}}{\left(3^{-x^2-y^2}+1\right)^2}$$



```python
def f(x,y):
    r=3**(-x**2-y**2)
    return 1/(r+1)

def fpx(x,y):
    r_1=3**(-x**2-y**2)
    r_2=(3**(-x**2-y**2)+1)**2
    return 2*x*np.log(3)*r_1/r_2

def fpy(x,y):
    r_1=3**(-x**2-y**2)
    r_2=(3**(-x**2-y**2)+1)**2
    return 2*y*np.log(3)*r_1/r_2
```

# Batch gradient descent with sympy

```python
gradient_x_slop=[]
gradient_y_slop=[]
mulip=0.1
param=np.array([1.8,1.0])
max_iter=300
values_par=param.reshape(1,2)
for n in range(max_iter):
    gradient_x=fpx(param[0],param[1])
    gradient_y=fpy(param[0],param[1])
    Gradient = np.array([gradient_x,gradient_y])
    param= param - mulip * Gradient
    gradient_x_slop.append(gradient_x)
    gradient_y_slop.append(gradient_y)
    values_par=np.append(arr=values_par,values=param.reshape(1,2),axis=0)
```

# MSE Mean squared error
## $$ \frac{1}{2m}\sum_{i=1}^{n} \left(y^{(i)}-h_{\theta}(x)^{(i)} \right)^2 $$

```python
#sample data
x_5 = np.array([[0.1,1.2,2.4,3.2,4.1,5.7,6.5 ]]).transpose()
y_5=np.array([[1.7,2.4,3.5,3.0,6.1,9.4,8.2]]).transpose()

regr=LinearRegression()
result=regr.fit(x_5,y_5)
b=result.coef_[0][0]
a=result.intercept_[0]
print("intercept or theta_0 by sklearn is: ",result.intercept_[0])
print("coeff or theta_1 by sklearn is: ",result.coef_[0][0])
y_hat = 0.8475351486029536 + 1.2227264637835915*x_5
```

```python
def mes(y_5,y_hat):
    r=1/len(y_5) *sum((y_5-y_hat)**2)
    return r
mes = mes(y_5,y_hat)
```

## Cost function MSE
## $$ \frac{\partial{MSE}}{\partial{\theta_{0}}}= -\frac{2}{n}\sum_{i=1}^{n} \big(y^{(i)} -\theta_{0} -\theta_{1}x^{(i)} \big) $$
## $$ \frac{\partial{MSE}}{\partial{\theta_{1}}}=-\frac{2}{n}\sum_{i=1}^{n} \big(y^{(i)} -\theta_{0} -\theta_{1}x^{(i)} \big)\big(x^{(i)}\big) $$



```python
nr_th=200
th_0=np.linspace(-1,3,nr_th)
th_1=np.linspace(-1,3,nr_th)
plot_t0,plot_t1=np.meshgrid(th_0,th_1)
plot_cost=np.zeros((nr_th,nr_th))
for i in range(nr_th):
    for j in range(nr_th):
        y_hat = plot_t0[i][j]+ plot_t1[i][j]*x_5
        plot_cost[i][j]=ms(y_5,y_hat)
plot_cost.shape
```

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size
    
    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    
    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier=0.01
thatas=np.array([2.8,2.0])
plot_vals=thatas.reshape(1,2)
mse_vals=ms(y_5,thatas[0]+thatas[1]*x_5)
for i in range(500):
    thatas=thatas-multiplier*grad(x_5,y_5,thatas)
    plot_vals=np.concatenate((plot_vals,thatas.reshape(1,2)))
    mse_vals=np.append(arr=mse_vals,values=ms(y_5,thatas[0]+thatas[1]*x_5))
```
